### Funzione RMS

```python
def rms(data, window):
    """ performs the moving-window smoothing of a signal using RMS """
    n = len(data)
    rms_signal = np.zeros((n-window)/(window/2)+1)
    rms_t = np.zeros((n-window)/(window/2)+1)
    i=0
    k=0
    while i<n-window:
        small_index = i
        big_index = (i+window)   # intended to avoid boundary effect
        window_samples = data[small_index:big_index]

        # here is the RMS of the window, being attributed to rms_signal 'k'th sample:
        rms_signal[k] = np.sqrt(sum([s**2 for s in window_samples])/len(window_samples))
        rms_t[k] = i+window/2
        i=i+window/2
        k=k+1
    return rms_signal,rms_t
```
